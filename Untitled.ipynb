# Encoding Categorical Columns

```python
le = LabelEncoder()

for col in ['Sub-Category', 'PaymentMode', 'State', 'City']:
    df[col] = le.fit_transform(df[col])
```

```python
X = df.drop(columns='Category', axis=1)
y = df['Category']
X_train,X_test, y_train,y_test = train_test_split(X,y, test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Models Building

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest Classifier': RandomForestClassifier(),
    'Bagging Classifier': BaggingClassifier(),
    'Adaboost Classifer': AdaBoostClassifier(),
    'Support Vector Machine (SVM)': SVC(),
    'Decision Tree Classifier': DecisionTreeClassifier(),
    'KNN': KNeighborsClassifier()
}
```

```python
results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    accuracy = accuracy_score(y_test, preds)
    pre = precision_score(y_test, preds, average='macro')
    recall = recall_score(y_test, preds, average='macro')
    
    print('\n -------------------------------------')
    print(f"\n✔ Model:  {name}")
    print(f"🚀 Accuracy: {accuracy}") 
    print(f"🧠 Precision: {pre}") 
    print(f"👉 Recall Score: {recall}") 
    
    confution_m = confusion_matrix(y_test, preds)
    result = ConfusionMatrixDisplay(confusion_matrix=confution_m)
    result.plot()
    
   
    
    results.append({
        'Model': name,
        'Accuracy': accuracy,
        'Precision': pre,
        'Recall': recall
    })
    results_df = pd.DataFrame(results)

# Show all model evaluation metrics in one table
print("\n📊 Model Evaluation Summary:")
results_df   
 
```
